```python
import socket
import threading

client = socket.socket()
client.connect(("localhost", 9997))

print(">> Connected to chat room")

def listen_messages():
    while True:
        try:
            incoming = client.recv(1024).decode()

            if not incoming or incoming.lower() == "exit":
                print(">> Server closed the connection")
                client.close()
                break

            print(incoming)

        except:
            break

threading.Thread(target=listen_messages).start()

while True:
    outgoing = input("You: ")
    client.send(outgoing.encode())

    if outgoing.lower() == "exit":
        print(">> You left the chat")
        client.close()
        break
```
