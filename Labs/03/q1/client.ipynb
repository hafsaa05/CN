```python
import socket

client_socket = socket.socket()
client_socket.connect(("localhost", 9998))

print(">> Connected successfully with server")

while True:
    choice = input("Choose: Message (M) or File (F): ").strip().lower()
    client_socket.send(choice.encode())

    if choice == "m":
        user_msg = input("You: ")
        client_socket.send(("MSG|" + user_msg).encode())

        if user_msg.lower() == "exit":
            print(">> Chat closed by client")
            break

        server_response = client_socket.recv(1024).decode()
        print("Server says:", server_response)

        if server_response.lower() == "exit":
            print(">> Server ended the chat")
            break

    elif choice == "f":
        file_name = input("Enter file name: ")
        client_socket.send(("FILE|" + file_name).encode())

        with open(file_name, "rb") as f:
            client_socket.send(f.read())

        print(">> File sent to server")

        server_response = client_socket.recv(1024).decode()
        print("Server says:", server_response)

        if server_response.lower() == "exit":
            print(">> Server closed connection")
            break

client_socket.close()
```
