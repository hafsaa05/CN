```python
import socket

server_socket = socket.socket()
server_socket.bind(("localhost", 9998))
server_socket.listen(1)

print(">> Server is running... waiting for client")

conn, address = server_socket.accept()
print(">> Client connected from:", address)

while True:
    client_choice = conn.recv(1024).decode()

    if client_choice == "m":
        data = conn.recv(1024).decode()

        if data.startswith("MSG|"):
            message = data.split("|", 1)[1]
            print("Client:", message)

            if message.lower() == "exit":
                print(">> Client exited chat")
                break

            server_msg = input("Reply: ")
            conn.send(server_msg.encode())

            if server_msg.lower() == "exit":
                print(">> Server closed chat")
                break

    elif client_choice == "f":
        file_info = conn.recv(1024).decode()

        if file_info.startswith("FILE|"):
            filename = file_info.split("|", 1)[1]
            print(">> Receiving file:", filename)

            file_bytes = conn.recv(4096)
            with open(filename, "wb") as f:
                f.write(file_bytes)

            print(">> File saved successfully")

            reply = input("Reply to client: ")
            conn.send(reply.encode())

            if reply.lower() == "exit":
                print(">> Server closed connection")
                break

conn.close()
server_socket.close()
```
